# 03 - Vectorización

Transformación de texto a representaciones numéricas: TF-IDF y Bag of Words.

**Objetivos:**
- Crear matriz TF-IDF
- Crear matriz Bag of Words
- Analizar features más importantes
- Guardar vectorizadores entrenados

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../')

from src.vectorizacion import crear_tfidf_vectorizer, crear_bow_vectorizer, get_top_features
import pickle
```

## 1. Cargar corpus procesado

```python
# Cargar corpus strings
df_procesado = pd.read_csv('../data/processed/corpus_procesado.csv')
documentos = df_procesado['opiniones_procesadas'].tolist()
print(f"Total de documentos: {len(documentos)}")
print(f"\nPrimer documento: {documentos[0]}")
```

## 2. Crear matriz TF-IDF

```python
# Crear vectorizador TF-IDF
matriz_tfidf, vectorizer_tfidf, features_tfidf = crear_tfidf_vectorizer(
    documentos,
    max_features=1000
)

print(f"Forma matriz TF-IDF: {matriz_tfidf.shape}")
print(f"Número de features: {len(features_tfidf)}")
```

## 3. Crear matriz Bag of Words

```python
# Crear vectorizador Bag of Words
matriz_bow, vectorizer_bow, features_bow = crear_bow_vectorizer(
    documentos,
    max_features=1000
)

print(f"Forma matriz BoW: {matriz_bow.shape}")
print(f"Número de features: {len(features_bow)}")
```

## 5. Guardar vectorizadores

```python
# Guardar vectorizadores entrenados
with open('../data/processed/vectorizer_tfidf.pkl', 'wb') as f:
    pickle.dump(vectorizer_tfidf, f)

with open('../data/processed/vectorizer_bow.pkl', 'wb') as f:
    pickle.dump(vectorizer_bow, f)

# Guardar matrices
import scipy.sparse as sp
sp.save_npz('../data/processed/matriz_tfidf.npz', matriz_tfidf)
sp.save_npz('../data/processed/matriz_bow.npz', matriz_bow)

print("✅ Vectorizadores guardados en data/processed/")
```
